```python
lora_8_log = "/mnt/nas-alinlp/zhuochen.zc/others/KnowB2/training_scripts/0505_qwen25_all_in_one_even.log"
lora_8_data = []
with open(lora_8_log) as f:
    lines = f.readlines()
    for line in lines:
        try:
            d = ast.literal_eval(line)
            lora_8_data.append(d)
        except:
            # print(line)
            pass
print(len(lora_8_data))

lora_8_loss = [i['loss'] for i in lora_8_data if 'loss' in i]
lora_8_token_acc = [i['token_acc'] for i in lora_8_data if 'token_acc' in i]
lora_8_eval_loss = [i['eval_loss'] for i in lora_8_data if 'eval_loss' in i]
lora_8_eval_token_acc = [i['eval_token_acc'] for i in lora_8_data if 'eval_token_acc' in i]
print(len(lora_8_loss), len(lora_8_token_acc), len(lora_8_eval_loss), len(lora_8_eval_token_acc))

lora_8_loss = lora_8_loss[:20]
lora_8_token_acc = lora_8_token_acc[:20]
lora_8_eval_loss = lora_8_eval_loss[:8]
lora_8_eval_token_acc = lora_8_eval_token_acc[:8]
```

```python
lora_32_log = "/mnt/nas-alinlp/zhuochen.zc/others/KnowB2/training_scripts/0526_qwen25_lora32_more.log"
lora_32_data = []
with open(lora_32_log) as f:
    lines = f.readlines()
    for line in lines:
        try:
            d = ast.literal_eval(line)
            lora_32_data.append(d)
        except:
            # print(line)
            pass
print(len(lora_32_data))

lora_32_loss = [i['loss'] for i in lora_32_data if 'loss' in i]
lora_32_token_acc = [i['token_acc'] for i in lora_32_data if 'token_acc' in i]
lora_32_eval_loss = [i['eval_loss'] for i in lora_32_data if 'eval_loss' in i]
lora_32_eval_token_acc = [i['eval_token_acc'] for i in lora_32_data if 'eval_token_acc' in i]
print(len(lora_32_loss), len(lora_32_token_acc), len(lora_32_eval_loss), len(lora_32_eval_token_acc))
```

```python
fft_log = "/mnt/nas-alinlp/zhuochen.zc/others/KnowB2/training_scripts/0526_qwen25_full_more.log"
fft_data = []
with open(fft_log) as f:
    lines = f.readlines()
    for line in lines:
        try:
            d = ast.literal_eval(line)
            fft_data.append(d)
        except:
            # print(line)
            pass
print(len(fft_data))

fft_loss = [i['loss'] for i in fft_data if 'loss' in i]
fft_token_acc = [i['token_acc'] for i in fft_data if 'token_acc' in i]
fft_eval_loss = [i['eval_loss'] for i in fft_data if 'eval_loss' in i]
fft_eval_token_acc = [i['eval_token_acc'] for i in fft_data if 'eval_token_acc' in i]
print(len(fft_loss), len(fft_token_acc), len(fft_eval_loss), len(fft_eval_token_acc))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# =============================================================================
# 1. 准备数据 (生成模拟数据)
# 在实际使用中，请将这些列表替换为您自己的真实数据。
# =============================================================================
# 为了模拟真实训练过程，Loss通常会下降，而Accuracy会上升。
np.random.seed(42) # 设置随机种子以保证结果可复现

# =============================================================================
# 2. 数据预处理
# =============================================================================
# 按要求，各取前20个数据点
num_points = 20
lora_8_loss_slice = lora_8_loss[:num_points]
lora_8_token_acc_slice = lora_8_token_acc[:num_points]
lora_32_loss_slice = lora_32_loss[:num_points]
lora_32_token_acc_slice = lora_32_token_acc[:num_points]
fft_loss_slice = fft_loss[:num_points]
fft_token_acc_slice = fft_token_acc[:num_points]

# 生成横轴 (X-axis) 数据，每个点跨越200个step
steps = np.arange(1, num_points + 1) * 200

# =============================================================================
# 3. 开始绘图
# =============================================================================
# 设置中文字体，以防显示乱码
# plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'Arial Unicode MS'] # 尝试使用系统支持的中文字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号'-'显示为方块的问题

# --- 可调整的美学参数 ---
FIG_SIZE = (12, 8)       # 图表整体尺寸
TITLE_FONT_SIZE = 40     # 标题字体大小
LABEL_FONT_SIZE = 28     # 坐标轴标签字体大小
TICK_FONT_SIZE = 24      # 坐标轴刻度字体大小
LEGEND_FONT_SIZE = 24    # 图例字体大小
LINE_WIDTH = 5         # 折线宽度
MARKER_SIZE = 13          # 数据点标记大小
GRID_ALPHA = 0.6         # 网格透明度

# # 定义颜色
# COLOR_L8 = '#1f77b4'     # 蓝色系 for LoRA r=8
# COLOR_L32 = '#2ca02c'    # 绿色系 for LoRA r=32
# COLOR_FFT = '#ff7f0e'    # 橙色系 for FFT

# # LoRA r=8 (蓝色系)
# COLOR_L8_LOSS = '#8ab4f7'   # 较浅的蓝色 for loss
# COLOR_L8_ACC = '#1f77b4'    # 较深的蓝色 for accuracy

# # LoRA r=32 (绿色系)
# COLOR_L32_LOSS = '#a0d9a0'  # 较浅的绿色 for loss
# COLOR_L32_ACC = '#2ca02c'   # 较深的绿色 for accuracy

# # FFT (橙色系)
# COLOR_FFT_LOSS = '#ffb574'  # 较浅的橙色 for loss
# COLOR_FFT_ACC = '#ff7f0e'   # 较深的橙色 for accuracy

# LoRA r=32
COLOR_L32_LOSS = '#BBDDDB'
COLOR_L32_ACC = '#529E9B'

# LoRA r=8
COLOR_L8_LOSS = '#F3D7B8'
COLOR_L8_ACC = '#E8AF72' 

# FFT
COLOR_FFT_LOSS = '#F1CCC6'
COLOR_FFT_ACC = '#D17061'


COLOR_LOSS = 'black'
COLOR_ACC = 'black'


# 创建图表和主要坐标轴 (ax1 for Loss)
fig, ax1 = plt.subplots(figsize=FIG_SIZE)

# --- 绘制 Loss 折线 (在 ax1 上) ---
# 使用实线表示 Loss
ax1.plot(steps, lora_8_loss_slice,  color=COLOR_L8_LOSS,  linestyle='-', marker='o', markersize=MARKER_SIZE, linewidth=LINE_WIDTH, label='LoRA (r=8) Loss')
ax1.plot(steps, lora_32_loss_slice, color=COLOR_L32_LOSS, linestyle='-', marker='s', markersize=MARKER_SIZE, linewidth=LINE_WIDTH, label='LoRA (r=32) Loss')
ax1.plot(steps, fft_loss_slice,     color=COLOR_FFT_LOSS, linestyle='-', marker='^', markersize=MARKER_SIZE, linewidth=LINE_WIDTH, label='FFT Loss')

# 设置 ax1 的坐标轴标签和刻度
ax1.set_xlabel('Step', fontsize=LABEL_FONT_SIZE)
ax1.set_ylabel('Train Loss', fontsize=LABEL_FONT_SIZE, color=COLOR_LOSS)
ax1.tick_params(axis='y', labelcolor=COLOR_LOSS, labelsize=TICK_FONT_SIZE)
ax1.tick_params(axis='x', labelsize=TICK_FONT_SIZE)


# 创建共享X轴的次要坐标轴 (ax2 for Token Accuracy)
ax2 = ax1.twinx()

# --- 绘制 Token Accuracy 折线 (在 ax2 上) ---
# 使用虚线表示 Accuracy，以作区分
ax2.plot(steps, lora_8_token_acc_slice,  color=COLOR_L8_ACC,  linestyle='--', marker='o', markersize=MARKER_SIZE, linewidth=LINE_WIDTH, label='LoRA (r=8) Token Acc')
ax2.plot(steps, lora_32_token_acc_slice, color=COLOR_L32_ACC, linestyle='--', marker='s', markersize=MARKER_SIZE, linewidth=LINE_WIDTH, label='LoRA (r=32) Token Acc')
ax2.plot(steps, fft_token_acc_slice,     color=COLOR_FFT_ACC, linestyle='--', marker='^', markersize=MARKER_SIZE, linewidth=LINE_WIDTH, label='FFT Token Acc')

# 设置 ax2 的坐标轴标签和刻度
ax2.set_ylabel('Train Token Accuracy', fontsize=LABEL_FONT_SIZE, color=COLOR_ACC)
ax2.tick_params(axis='y', labelcolor=COLOR_ACC, labelsize=TICK_FONT_SIZE)

# --- 设置图表整体元素 ---
# 标题
# plt.title('Training Curve', fontsize=TITLE_FONT_SIZE, pad=20)

# 网格
ax1.grid(True, linestyle='--', alpha=GRID_ALPHA)

# 图例 (合并两个坐标轴的图例)
# 从ax1和ax2获取句柄(handles)和标签(labels)
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
# 将它们合并并显示在一个图例中
# ax1.legend(h1 + h2, l1 + l2, loc='center right', fontsize=LEGEND_FONT_SIZE, bbox_to_anchor=(1, 0.5))

ax1.legend(h1, l1, loc='lower center', fontsize=LEGEND_FONT_SIZE, bbox_to_anchor=(0.5,0))
ax2.legend(h2, l2, loc='center right', fontsize=LEGEND_FONT_SIZE, bbox_to_anchor=(1,0.6))

# 优化布局，防止标签重叠
plt.tight_layout()

# 保存图表（可选）
# plt.savefig("training_curve.png", dpi=300, bbox_inches='tight')

# 显示图表
plt.show()
```
